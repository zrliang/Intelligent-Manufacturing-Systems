```python
# import matplotlib相關套件

import matplotlib.pyplot as plt

rb=0.375
ct=41
W0=15.375

WIP = [i+1 for i in range(25)]
CT=[]
for w in WIP:
    if w <= W0:
        ct_b=ct
    else:
        ct_b=round(w/rb,1)
    CT.append(ct_b)


plt.plot(WIP,CT,'s-',color = 'r')

#X,Y軸
plt.xlabel("WIP", fontsize=30)
plt.ylabel("CT", fontsize=30)

#數字標籤
for a, b in zip(WIP, CT):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=10)
# 畫出圖片

plt.show()
```

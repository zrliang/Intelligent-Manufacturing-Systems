```python
import matplotlib.pyplot as plt

rb=0.4
T0=20
W0=8

# CT_vs_WIP
## CT_Best

WIP = [i for i in range(27) if i % 2 == 0 and i !=0] #2-26 偶
CT=[]
for w in WIP:
    if w <= W0:
        ct_b=T0
    else:
        ct_b=round(w/rb,1)
    CT.append(ct_b)

plt.plot(WIP,CT,'s-',color = 'r')

#X,Y軸
plt.xlabel("WIP", fontsize=30)
plt.ylabel("CT", fontsize=30)

#數字標籤
for a, b in zip(WIP, CT):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=10)
# 畫出圖片

plt.show()
```

```python
import numpy as np                                        # 引入科學計算函式庫
import matplotlib.pyplot as plt                           # 引入繪圖函式庫

xmin, xmax, num = 1, 30, 100                              # 設定繪圖範圍、取點數
w = np.linspace(xmin, xmax, num)                          # 產生x

plt.figure(figsize = (6, 4.5), dpi = 100)                 # 設定圖片尺寸
plt.xlabel("WIP", fontsize = 16)                          # 設定坐標軸標籤
plt.ylabel("CT",  fontsize = 16) 
plt.xticks(fontsize = 12)                                 # 設定坐標軸數字格式
plt.yticks(fontsize = 12)
plt.grid(color = 'red', linestyle = '--', linewidth = 1)  # 設定格線顏色、種類、寬度
plt.ylim(0, 80)                                          # 設定y軸繪圖範圍

rb=0.4
T0=20
W0=8

## CT_Best
WIP = [i+1 for i in range(27)] 
CT=[]
for w_b in WIP:
    if w_b <= W0:
        ct_b=T0
    else:
        ct_b=round(w_b/rb,1)
    CT.append(ct_b)

# CT_Worst
CT_Worst = 20*w

# PWC
CT_PWC = T0+(w-1)/rb
                                  

# 繪圖並設定線條顏色、寬度、圖例
plt.plot(w, CT_Worst, color = 'red', linewidth = 3, label = 'Worst Case')    
plt.plot(w, CT_PWC, color = 'Green', linewidth = 3, label = 'Practical Worst Case')         
plt.plot(WIP, CT, color = 'blue', linewidth = 3, label = 'Best Case')


plt.legend(loc='lower right') # 圖例
#plt.savefig('Fe_r_plot.svg')                              # 儲存圖片
#plt.savefig('Fe_r_plot.png')
plt.show()    
```
